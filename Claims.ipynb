**● How does the number of claims differ across locations?**

Totally 98 claims were come from 4 different locations. The highest number of claims happened from Sao Luis, which reach 30 cases in total. Recife had 25 claim cases which ranked the second claims number among the four locations. Fortaleza and Natal had 22 and 21 cases respectively which became third and fourth place in terms of the claims number. In general, all the four locations had  20-30 cases, that is relatively close.

**● What is the distribution of time to close claims?**

First, we focus on is there any outlier of those time to close values. We created 8 time range groups from 0 to 3500 days, and the range size is 500. Then we noticed that there might be one outlier belonging 3500 group, and no case in 3000 group, which seems far from other cases. However, when we look at the detail, the longest time to close was 3591 days and the second was 2997 days, which were in the reasonable claim period. So we decided there is no outlier among these values.

![Q2-1](./Q2-1.png)

Next, we turn to analyse the distribution of time to close claims. As we can see from the graph, the largest number of claim (37 cases) had been close claims within 500 days and 30 cases had been finished in 1000 days. These two large groups had been accounted for around 68% of total cases, which completed in the regular time frame. On the contrary, there are 8 cases that had been done more than 2000 days including 2 abnormal cases which completed over 2500 days. We suggest these cases need to be figure out why took so long and is there any optimisation of process for these issues.
